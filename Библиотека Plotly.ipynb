```python
import pandas as pd
import seaborn as sns
import sqlite3
import matplotlib.pyplot as plt
```

```python
conn = sqlite3.connect('data\checking-logs.sqlite')
```

```python
checker = pd.read_sql('''SELECT * FROM checker''', conn)
checker
```

```python
num_commits = pd.read_sql('''WITH num_commits (uid, timestamp) AS (SELECT uid, date(timestamp) FROM checker 
                             WHERE uid LIKE "user_%" AND labname = "project1" AND status = "ready") 
                             SELECT uid, timestamp, COUNT(*) AS numTrials FROM num_commits GROUP BY uid, timestamp''', conn)
num_commits
```

```python
num_commits = num_commits.sort_values(['timestamp', 'uid']).pivot(index='uid', columns='timestamp', values='numTrials').fillna(0).cumsum(axis=1)
num_commits
```

```python
num_commits = num_commits.transpose().sort_index().transpose()
num_commits.reset_index(inplace=True)
num_commits
```

```python
import plotly
import plotly.graph_objects as go
import plotly.express as px
import plotly.io as pio
import numpy as np
```

```python
numOfRows = num_commits.shape[0] 
numOfCols = num_commits.shape[1] 
numOfFrames = numOfCols - 1
xaxis_range = [0,numOfFrames + 2]
x_init = np.array([1])
```

```python
max_commits = num_commits.drop(columns='uid').max().max()
```

```python
initial_data = []
for user_ind in range(numOfRows):
    y_axis = np.array(num_commits.iloc[user_ind,0])
    initial_data.append(go.Scatter(x =x_init, y = y_axis,mode = "lines",name = num_commits.uid[user_ind]))
initial_max = max_commits
```

```python
frames = []
for f in range(1,numOfFrames+1):
    x_axis = np.arange(1,f+1)
    curr_data = []
    title_names = []
    for user_ind in range(27):
        user = num_commits.uid[user_ind]
        y_axis = np.array(num_commits.iloc[user_ind,1:f+1])
        curr_data.append(go.Scatter(x = x_axis, y = y_axis,mode = "lines+markers", name = user))
    curr_frame = go.Frame(data = curr_data)
    frames.append(curr_frame)
```

```python
figure = go.Figure(
    data = initial_data,
    layout = {
        "title":"Dynamic of commits per user in project1",
        "xaxis":{"range":xaxis_range, "visible":True, "showline":True},
        "yaxis":{"visible":True, "showline":True},
        "updatemenus":[{"type":"buttons","buttons":[{"method":"animate","label":"play", "args":[None]}]}],
        "width":1200,
        "height":500
        },
    frames = frames,layout_yaxis_range=[0,max_commits]
    )
figure.show()
```
